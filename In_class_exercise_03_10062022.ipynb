Question 2 (20 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
# You code here (Please add comments in the code):
from six.moves import urllib
import pandas as pd
from urllib.request import urlopen


text = []
url = "https://raw.githubusercontent.com/chaitu1121/SAICHAITANYA_INFO5731_FALL2022/main/readme.txt"
x= urllib.request.urlopen(url)  

for i in x:
    dec_data = i.decode("utf-8").replace('\r\n', '')   
    if dec_data: 
        text.append(dec_data)       #appending the decoded data to a list
data = pd.DataFrame (text, columns = ['Text'])   #converting the list to data frame using pamdas
print(data)
```

Question 3 (10 points): Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)." Select the most important features you extracted above, rank the features based on their importance in the descending order. 

```python

# You code here (Please add comments in the code):
#to find num of sentences and words 

data['Num of sentences'] = data['Text'].apply(lambda y: len(str(y).split(".")))
data['Num of Words'] = data['Text'].apply(lambda y: len(str(y).split(" ")))
data['Num of Char'] = data['Text'].str.len()

data
```

```python
def avg(x):
    jai = x.split() 
    if len(jai) != 0:  
        return(sum(len(deep) for deep in jai)/len(jai))
    else:
        return None

data['Avg Word Length'] = data['Text'].apply(lambda z: avg(z))
data
```
