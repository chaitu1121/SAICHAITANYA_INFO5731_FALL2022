```python

import requests
import json
from bs4 import BeautifulSoup
import csv

```

```python
# specify the url
url = "https://www.emerald.com/insight/publication/issn/0264-0473"

# query the website and return the html to the variable â€˜pageâ€™
page = requests.get(url)
 
# parse the html using beautiful soup and store in variable `soup`
soup = BeautifulSoup(page.content, 'html.parser')

# Get the volume and issue information
volumes = soup.find_all('div', class_="container content_block")

volumes
```

```python
# Create csv file
csv_file = open('emerald-scrape.csv', 'w') 
csv_writer = csv.writer(csv_file) 
csv_writer.writerow(['Authors', 'Author Institutions', 'Year', 'Abstract', 'Keywords', 'Doi', 'References', 'Citations'])
```

```python
# Iterate over each volume
for volume in volumes:
    issues = volume.find_all('div', class_="issue-item")
    # Iterate over each issue
    for issue in issues:
        # Get each paper's author and institution
        articles = issue.find_all('div', class_="article-item")
        for article in articles:
            authors = article.find_all('div', class_="author")
            # Get the authors and their institutions
            author_list = []
            institution_list = []
            for author in authors:
                author_name = author.find('span', class_="author-name").text
                author_list.append(author_name)
                institution = author.find('span', class_="author-institution").text
                institution_list.append(institution)
            # Get the year, abstract, and keywords
            year = article.find('span', class_="publication-year").text
            abstract = article.find('span', class_="abstract-text").text
            keywords = article.find('span', class_="keywords-text").text
            # Get the doi
            doi = article.find('span', class_="doi-text").text
            # Get the references and citations
            url = 'https://api.semanticscholar.org/v1/paper/' + doi
            response = requests.get(url)
            data = response.json()
            references = data['references']
            citations = data['citations']
            # Write all the data to the csv file
            csv_writer.writerow([author_list, institution_list, year, abstract, keywords, doi, references, citations])

# Close the csv file
csv_file.close()
```
