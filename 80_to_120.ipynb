```python
from requests import get
from bs4 import BeautifulSoup
from warnings import warn
from time import sleep
from random import randint
import numpy as np, pandas as pd
import requests
```

```python
url='https://www.emerald.com/insight/publication/issn/0264-0473'
page=requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
vols=soup.find_all('a')
b=[]
a=[]
```

```python
for x in vols:
    if 'publication' and 'issn' in x['href']:
        vol_url=''
        if 'https' not in x['href']:
                vol_url='https://www.emerald.com'+ x['href']
        else:
            vol_url=x['href']
        b.append(vol_url)
```

```python
b.reverse()
b
```

```python
len(b)
```

```python
for i in b[80:120]:
    p=requests.get(i)
    s = BeautifulSoup(p.text, 'html.parser')
    link=s.find_all('a')
    for l in link:
        if 'full' and 'html' in l['href']:
            if 'https'  not in l['href']:
                l_url='https://www.emerald.com' + l['href']
            else:
                l_url=l['href']
            a.append(l_url)
    
    
    
```

```python

for x in a:
    p=requests.get(x)
    sp=BeautifulSoup(p.text,'lxml')
    k=sp.find('section',{'class':'intent_sub_content Abstract__block__text'})
    g.append(k)  
        
        

            
```

```python
for d in a:
    p=requests.get(d)
    sg=BeautifulSoup(p.text,'lxml')
    for row in sg.find_all('h1'):
        title.append(row.text)
    
```

```python

for x in a:
    p=requests.get(x)
    sp=BeautifulSoup(p.text,'lxml')
    k=sp.find_all('a',{'class':'intent_doi_link Citation__identifier__link'})
    doi.append(k)  
        
        

            
```

```python

for x in a:
    p=requests.get(x)
    sp=BeautifulSoup(p.text,'lxml')
    k=sp.find_all('span',{'class':'intent_journal_publication_date'})
    date.append(k)  
        
        

            
```

```python
d=pd.DataFrame({'BookName':title,'Doi':doi,'Date':date,'Abstract':g})
d
```

```python
from IPython.display import HTML
import base64  
import pandas as pd  

def create_download_link( df, title = "Download CSV file", filename = "data.csv"):  
    csv = df.to_csv(index =False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(d)
```
